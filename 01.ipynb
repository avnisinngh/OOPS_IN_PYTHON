```python
class Student:
    def __init__(self,name, rollNumber, marks): # parameters
        self.name = name
        self.rollNumber = rollNumber
        self.marks = marks


s1 = Student("Avni", 12, 90) # arguments
print(s1.marks)
```

# Methods

```python
class Student:
    def __init__(self,name, rollNumber, marks): # parameters
        self.name = name
        self.rollNumber = rollNumber
        self.marks = marks

    def study(self):
        print("I am Studying")

    def play(self):
        print("I am going to play")


s1 = Student("Avni", 12, 90) # arguments
print(s1.marks)

s1.play()
```
